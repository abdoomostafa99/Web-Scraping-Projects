```python
import requests
import csv
import pandas as pd
from bs4 import BeautifulSoup
url = 'https://diwanegypt.com/product-category/books/arabic-books/page/1/'
response = requests.get(url)
soup = BeautifulSoup(response.content , 'html')
```

```python
all_books = soup.find_all('li' , {'class' :'product'})
```

```python
data = []
```

```python
for i in range(1,150):
    url = f'https://diwanegypt.com/product-category/books/arabic-books/page/{str(i)}/'
    response = requests.get(url)
    soup = BeautifulSoup(response.content , 'lxml')
    all_books = soup.find_all('li' , {'class' :'product'})
    
    for book in all_books:
        name_of_book = book.find('h2' , {'class' : 'woocommerce-loop-product__title'}).text
        price = book.find('bdi').text
        dicto = {"book name" : name_of_book , 'price' : price}
        data.append(dicto)
```

```python
df = pd.DataFrame(data)
df
```

```python
df.to_csv('diwan_book.csv' , index = False )
```

```python
with open("diwan_books.csv","w",encoding="UTF-8",newline='') as fd:
    heads = ["book name",'price']
    writer = csv.DictWriter(fd,heads)
    writer.writeheader()
    
    for i in range(2):
        url = f'https://diwanegypt.com/product-category/books/arabic-books/page/{str(i)}/'
        response = requests.get(url)
        soup = BeautifulSoup(response.content , 'lxml')
        all_books = soup.find_all('li' , {'class' :'product'})
        
        for book in all_books:
            name_of_book = book.find('h2' , {'class' : 'woocommerce-loop-product__title'}).text
            price = book.find('bdi').text
            dicto = {"book name" : name_of_book , 'price' : price}
            writer.writerow(dicto)
```

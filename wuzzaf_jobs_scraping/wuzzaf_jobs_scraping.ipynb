```python
import requests
import csv
import pandas as pd
from bs4 import BeautifulSoup

with open('wuzzaf jops.csv' , 'w' , newline='' , encoding='UTF-8')as fd:
    fieldnames = ['jop name' , 'laocation' , 'company name' , 'experience' , 'experience year' , 'skills']
    writer = csv.DictWriter(fd , fieldnames = fieldnames)
    writer.writeheader()
    for page in range (48):  
        url = f'https://wuzzuf.net/search/jobs/?a=hpb%7Cspbg&q=Data%20Scince&start={page}'
        response = requests.get(url)
        soup = BeautifulSoup(response.content , 'lxml')
        jops_title = soup.find_all('div' , {'class' : 'css-pkv5jc'})
        
        for jopp_title in jops_title:
            name_of_jop = jopp_title.find('h2' , {'class' : 'css-m604qf'}).text

            location = jopp_title.find('span' , {'class' : 'css-5wys0k'}).text

            company_name = jopp_title.find('a' , {'class' : 'css-17s97q8'}).text.strip('- ')

            exp = jopp_title.find('div' , {'class' : 'css-y4udm8'}).find('a' , {'class' : 'css-o171kl'}).text

            if exp != 'Not specified':
                exp_year = jopp_title.find('div' , {'class' : 'css-y4udm8'}).find('a' , {'class' : 'css-o171kl'}).find_next().text[1:].strip()
            else :
                exp_year = 'None'

            listo = []
            skills = jopp_title.find('div' , {'class' : 'css-y4udm8'}).find('a' , {'class' : 'css-o171kl'}).find_next().find_next_siblings()
            for skill in skills:
                listo.append(''.join(skill.text.split('.css-5x9pm1{-webkit-text-decoration:none;text-decoration:none;color:inherit;max-width:calc(100vw / 2 - 32px);white-space:nowrap;overflow:hidden;text-overflow:ellipsis;display:-webkit-inline-box;display:-webkit-inline-flex;display:-ms-inline-flexbox;display:inline-flex;}')))
            writer.writerow({"jop name" : name_of_jop , 'laocation' : location , 'company name' : company_name , 'experience' : exp , 'experience year' : exp_year , 'skills' : listo})
```

```python
import requests
import csv
import pandas as pd
from bs4 import BeautifulSoup
date = input("Enter date of matches please for example MM/DD/YYYY")
url = f'https://www.yallakora.com/match-center/%D9%85%D8%B1%D9%83%D8%B2-%D8%A7%D9%84%D9%85%D8%A8%D8%A7%D8%B1%D9%8A%D8%A7%D8%AA?date={date}#'
response = requests.get(url)
soup = BeautifulSoup(response.content , 'lxml')
championships = soup.find_all('div' , {'class' : 'matchCard'})

with open('matches.csv' , 'w' , newline='' , encoding='UTF-8')as fd:
    fieldnames = ['champion' , 'teamA' , 'teamB' , 'match_res']
    writer = csv.DictWriter(fd , fieldnames = fieldnames)
    writer.writeheader()

    for championship in championships:
        dawry_name = championship.find('h2').get_text().strip()
        team1 = championship.find_all('div' , {'class' : 'teamA'})
        team2 = championship.find_all('div' , {'class' : 'teamB'})
        result = championship.find_all('div' , {'class' : 'MResult'})

        for i in range(len(team1)):
            team_1 = team1[i].find('p').get_text().strip("\t")
            team_2 = team2[i].find('p').get_text().strip("\t")
            match_res = f"{result[i].find_all('span' , {'class' : 'score'})[0].get_text()} - {result[i].find_all('span' , {'class' : 'score'})[1].get_text()}"
            writer.writerow({"champion" : dawry_name , 'teamA' : team_1 , 'teamB' : team_2 , 'match_res' : match_res})
```
